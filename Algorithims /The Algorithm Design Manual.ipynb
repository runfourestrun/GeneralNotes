### What is an Algo

* An algorithm is a procedure to accomplish a specific task
* An Algorthim is the idea behind any computer program
* To be interesting an Algorithm must solve a general well specified problem. 
* Determing you are dealing with a general problem is fundamental to solving it
* An algorithm takes any amount of desired input instances and transforms them to an output
