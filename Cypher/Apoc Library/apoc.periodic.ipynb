# apoc.periodic



### apoc.periodic.cancel
***

##### Description: cancel job with given name


##### usage example: canceling a job that creates people: 

```
CALL apoc.periodic.cancel('create-people')
```




### apoc.periodic.commit
*** 
##### Description:
runs the given statmenet in seperate transactions until it returns 0


```
WITH ["London", "Manchester", "Cardiff", "Birmingham", "Coventry", "Edinburgh"] AS cities
UNWIND range(1, 10000) AS id
MERGE (p:Person {id: id})
WITH cities, p, toInteger(rand() * size(cities)) AS index
SET p.city = cities[index];

CALL apoc.periodic.commit(
  "MATCH (person:Person)
   WHERE exists(person.city)
   WITH person limit (moneysign) limit
   MERGE (city:City {name:person.city})
   MERGE (person)-[:LIVES_IN]->(city)
   REMOVE person.city
   RETURN count(*)",
  {limit:1000});
  
```



### apoc.periodic.iterate
*** 

##### description: 

takes two queries/statements. The first statement is the data to be operated on (literals, nodes,relationships,properties,maps, rows of a csv file, etc.).  Second statement is the operation you want to create" 




#### Examples

```
CALL apoc.periodic.iterate('MATCH (p:Person) RETURN p','DETACH DELETE (p)', {batchSize:50,parrallel:True})
```
