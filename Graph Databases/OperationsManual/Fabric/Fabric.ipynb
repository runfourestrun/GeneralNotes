#  Neo4j Fabric
source: https://neo4j.com/docs/operations-manual/current/fabric/introduction/

### Overview


* Fabric introduced in Neo4j 4.0, is a way to store and retrieve data in multiple databases, whether they are on the same Neo4j DBMS or multiple DBMS, using a single Cypher query


#### Objectives

1. unified view of local and distributed data via a single client connection and session.
2. increased scalability for read/write operations , data volume and concurrency
3. predictable response time for queries executed during normal operations
4. High availability and no single point of failure. 


In practical terms...

1. Data Federation - Basically connecting disjointed graphs
2. Data Sharding - the ability to access data avilabile in distributes sources in the form of a common graph parition on multiple databases (?)




### 2 Concepts 


#### 2.1 the Fabric Database

* Fabric setup includes a ***Fabric virtual Database***, which acts as an entry point to a federated or sharded graph.
* This database is the execution context in which multi-graph queries can be executed. 
* Drivers and client application access and use this execution context by naming it as the selected database for the session.
* This virtual database stores no data, just relays it from elsewhere
* This Fabric virtual database can be configured on a standalone Neo4j only. 


#### 2.2 Fabric Graphs

* in Fabric virtual db, data is organized as a graph
* Graphs are seen by the client applications as local logical structures
* Databases can be local (on the same machine as Fabric Virtual db) or remote




### 3 Fabric Deployment Examples


#### Overview

* Consitutes an extremely versatile environment that provides scalability and availability with no single point of failure. 


### 3.1 Development Deployment


* In it's simplest deployment, Fabric can be used on a single instance, where Fabric graphs are associated to local databases. This approach is used by Software developers who want to create an application with multiple local Neo4j DBMS, or by power users who intend to execute Cypher queries against local, disjointed graphs. 




### 3.2 Cluster Deployment (no single point of failure) 

* In this deployment mode, Fabric guarentees access to disjoint graphs in high availability with no single point of failure. 
* Availability created through redundant entry points for the Fabric Database (multiple single instance Neo4j DBMSs with the same Fabric configuration) and a minimal Causal cluster.
