# IAM: Users & Groups


* Create users and assign them to groups
* Root account created by default
* Users are people within your organization, and can be grouped 
* Users can be added to Groups
* User can be added to multiple groups
* groups can not belong to groups
* Users or Groups can be assigned JSON documents as policies 





### IAM: Permissions



* Policies define the permissions of the users
* In AWS you apply the least privledge policy - don't give users more 



### IAM Inheritence 

* Users inherit policies from groups


### IAM policy structure



* Contains the following:

* Version: policy language
* ID: an identifier for the policy
* Statement: one or more individual statements (required)

* Statements consist of:

* Sid: an idenfier for the statement
* Effect: whether the statment allows access or not
* Principal: Account/user/role to which this policy is applied
* Action: list of actions this policy allows or denies
* Resource: list of resurces to which the actions applied to
* Condition: conditions for when the policy is in effect

#  Access to AWS


* To access AWS you have three options:

1. AWS management console (protected by password + MFA)
2. AWS command line interface (CLI): protected by access keys
3. AWS Software Development Kit: for code - protected by access keys


*Users manage their own access keys
*Treat Access key ID and Secret Access Key as CONFIDENTIAL 



### AWS CLI

* Direct access to the public APIs of AWS services
* You can develop scripts to manage your resources


### AWS SDK

* Language specific APIs
* Enables you to access and manage AWS services programmatically
* embedded within your application



# IAM best Pracitces

* Don't use root account other than setup
* One physical user = One AWS User
* Assign user to groups/ manage users through groups
* Create and use Roles for giving permissions to AWS services 
* Use access keys for AWS CLI/SDK
* Audit permissions through IAM credentials report
