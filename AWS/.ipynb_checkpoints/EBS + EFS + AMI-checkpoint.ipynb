# EBS - Elastic Block Storage



* Basically a network drive
* default storage for an ec2 instance
* needs to be in the same availability zone as your ec2 instance
* you can snapshot EBS volumes
* Can select IOPS & Size



### EBS Volumes

* EBS volumes are charachterized by:

1. Size
2. Throughput
3. IOPS

### EBS Encryption

When you encrypt a volume you get the following

* Data at rest is encrypted inside the volume
* all the data in flight moving between the instance and the volume is encrypted
* all snapshots are encryped 
* All volumes created from the snapshot 


You can encrypt an unecrypted volume by copying it. 


#### Volume Types

(only first two can be used as boot volumes)

1. Gp2/Gp3 : General Purpose SSD volumes that balances price and performance
2. io1 / io2: Highest performance SSD volume for mission critical low latency or high throughput workloads
3. st1 (Throughput optimized HDD): Low cost HDD volume for frequently accessed, throughput intensive workloads
4. Sc 1 (cold HDD) : Lowest cost HDD volume for less frequently accessed workloads




### Provisioned IOPS

* Great for database workloads
* Can increase the PIOPS independent from storage size
* Io2 is the best option
* I02 block express 
* Supports EBS Multi-attach





## EBS Snapshot

* Also known as a backup
* Not necessary to detach before snapshotting, but best practice
* Can transfer snapshots across region or AZ
* Basically if you want to do that you have to create a snapshot, than make another copy of that snapshot in the AZ you want


#### Workflow

* Create volume (or have it created when you spin up the Ec2 instance) -> Snapshot Volume -> Copy Snapshot





# AMI 

* AMI = Amazon Machine Image
* AMI are a customization of EC2 instance
 - You add your own software, configuration, operating system, monitoring .. 
 - faster boot / config time because all software is prepackaged
 
* AMI are built for specific regions
* You can launch EC2 instances from: 
1. A public AMI: AWS provixed
2. Your own AMI
3. An AWS Marketplace AMI: an AMI someone else made 






### EC2 Instance Store


* EBS Volume are network drives with good, but limited performance
* If you need a high performance hardware disk, use an EC2 Instance Store


* Better I/O performance
* EC2 Instance store lose their storage if they're stopped (ephemeral)
* Better for buffer, cache, scratch data, temporary data
* Backups and Replication are your responsibilitiy







# EFS - Elastic File System



* Managed NFS (network file system) that can be mounted on many EC2 instances
* Multi - AZ (different from EBS)
* Highly available, scalable, expensive (3x the cost of gp2 drive), pay per use




## EFS use cases:

* Content management, web serving, data sharing, Wordpress
* Uses NFSv4 protocol
* Uses security group to control access
* Only Linux
* Encryption at rest through KMS



### EFS Scale

* 1000s of concurrent NFS clients, 10 GB+ /s throughput
* Grow to Petabyte scale


### Performance mode (set at EFS creation)

* Gen Purpose: latency sensitive use cases (web server, CMS)
* Max I/O - higher latency, throughput, highly parrellal (big data)

### Throughput mode

* bursting (1TB = 50 MIbs + burst of up to 100 MIB/s)
* Provisioned - Set throughput regardless of instance size
