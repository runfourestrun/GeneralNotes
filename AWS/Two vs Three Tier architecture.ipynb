### Two vs Three Tier Architecture
Source: https://medium.com/@gacheruevans0/2-tier-vs-3-tier-architecture-26db56fe7e9c

### What is software architecture?

* Refers to the high level structures of a software system and the discipline of making such structures. Our focus, however, will be on these two structures: Two vs Three Tiered Applications




## Two Tier



#### High level

* Where you have a direct communication between a client and a server with no intermediary. 
* Two Central components are as follows:

1) Client Application
2) Database

* Client in a two tiered architecture has the code written for saving data in the database. 
* The Client sends a request to the server, where it then processes the request and sends back the data. 

Client handles both:

1) Presentation Layer (or application interface)
2) Application layer (or logical operations)

Server handles:

1) Database operations





## Three Tier




#### High level

* Three central components are as follows:

1) Presentation Layer (Client Tier) ............ public
2) Application Layer (Business Tier) .............. private
3) Database Layer (Data Tier) .............. private 

(private layers are only accessible in the network)

#### Presentation Layer (or client tier)
* The Presentation layer is the top most layer of an application. This is the layer seen when using the software (Interface, web pages). Its main function is to communicate with the application layer. 
* This layer passes inputs (mouse clicks, keyboard actions) to the application layer

#### Application Layer (or business logic tier)
* Here is where we fine logic controls and functionality that processes fata recieved from the presentation layer and database layer. It acts as an intermediary between the presentation and database layer.

#### DataBase Layer (or the Data Tier)
* Layer that stores data using storage and execution methods defined in the application layer. 



### Benefits of three-tier architecture

* Scalability - Each tier can scale horizotally. You can add servers with a Load Balancer (for example three servers) to handle more requests without scaling the other two layers
* Performance - Because the Presentation tier can cache requests, network utilization is minimized, and the load is reduced on the Application and Data Layers. IMPORTANT: You can load balance any tier. NOTE: probably should read more into how caching of requests work. 
* Availability - If application tier server is down and caching is sufficent, the presentation layer can process requests via cache


### Drawbacks of three-tier architecture
