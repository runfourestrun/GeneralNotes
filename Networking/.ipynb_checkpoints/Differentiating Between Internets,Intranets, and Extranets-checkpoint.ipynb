## Understanding a network


### what is a network?

#### A network can be defined as two end point devices connected by a medium. 

Can be...

* Severs
* Tablets
* Phones
* Cars
* Fridges


* What do all these devices have in common? They possess at least one network interace card (NIC), and they are running a network operating system (NOS) and have been coded in a way to take advantage of one or more network protocols
* A protocol refers to a set of rules or standards that must be adhered too
* One definition of a network is that it is a shared medium. Can be thought of as a cable over airways in the case of a wireless network. 
* identifying devices across a network require using hostnames, IP address and MAC address. We will discuss these in more detail later in this book. 


### Understanding the internet


* subtle difference between an internet and the Internet. 
* internet means the connection between different networks
* The Internet refers to services that are publicly available outside of your own networks and the supporting infrastructure. 
* Company a - Company b = internet
* connecting to facebook = The Internet (publicly available)



### Understanding intanets


* intra means inside
* relates to network communication within our network
* Access to these services will be controlled through user/device authentication to ensure that they remain private and inaccessible from the internet. 
* You will find that some organizations host internal web pages, such as the collaborative tool SharePoint


### Understanding extranet

* You can allow access to other entities (businesses, partners) and limited to only specific resources. 
* Assume company A needs to sell car engines, But they rely on Company B for a critical component of the engine, the pistons. Now they could hit up Company B on the phone, or Company B could provide them access via extranet. 





### Introduction to basic security features


* Firewalls
* DMZ 
* VPN


### Firewalls

* designed to allow or deny network traffic based upon a set of defined criteria
* rules are often referred to as access control enteries (ACEs), and a group of them form an ACL (access control list)
* rules are applied bi-directionally
* firewalls iterate through their rules
* You can apply firewalls on the host or network wide


### DMZ 

* A DMZ is a perimiter network. Allows the public to access certain network services while still mainting the security of your internal devices. 
* Basically an extranet, for the public
* Common services in include a web server and a Domain name Server




### VPN

* a means of transmitting private data securely from one network across an unsecured network to a third network. Generally speaking, the unsecured network we are referring to is the Internet, which, due to the nature of it's design, has a number of potential security risks. 
* You can install a VPN client on each device, remote access VPN
* Basically requires an authentication into the network, supposedly a pain in the ass to administer.
* VPN tunnells provide data security through the use of encryption and authentication. The methods that are used vary depending on the tunneling protocol thats used. PPTP,PPP, SSTP
* SSTP uses SSL/TLS to provite encryption.
* It also utilizes the same nework ports as HTTPS traffic, that is, port 443

 
 
 
 
 
 
 



# Understanding LAN (local area network)


### Scale of networks


* networks can be described in terms of their scale of deployment according to how much of an area they cover or what they cover. Basically networks are like an onion 





### Intro to LANS


#### LAN is something that covers a small geographic area, devices are local to each other
#### all of the devices are eitehr owned by me, or my organization


* Identifying devices on a local network can be done using hostnames, MAC addresses and IP adrress, all of which are unique in a LAN



#### Hostname

* A computers hostname is an easy to read method of identifying a device on a network. 
* hostname is configured by system admin
* When referring to a destination device by its hostname, the sending device will need to resolve it to an IP address.


#### IP addresses

* Hostnames are human friendly but not machine friendly
* Hostnames on their own would generally thought of as being local
* Therefore we need to resolve any hostnames to an IP address
* IPV4 address has two elements: a network element and a host element
* network element identifys network the device is on
* host element identifys the device
* network element is street name and host element is house number


#### limits on IPV4 addresses worldwide, we can reuse IP addresses across different LANs by following the RFC 1918
* allocates three ranges of IP addresses that can be used in LANs
* these addresses are private IP addresses




* When a transmitting device identified the IP address of a recipient device, it asks is it in the same network or a remote network



#### MAC address 

* A means of identifying a device on the local area network. It is an address that has many names. As well as being referred to as a MAC address, it is also referred to as a burnt in address.

## LANS using switches

* Switches have become the backbone of most orgs
* Switches allow for greater network segmentation but also offer full duplex communication 
* Switch basically creates a virtual connection between two devices.
* Does this by a source device sending data to a switch, the switch looks up the recieving devices port associated with it's MAC address and it sends data to that port. 
* Big benefit is all devices communicate at the same time. 
* Switch buffers data until there is a gap in the traffic, and then forwards it on. 
* This means that every port on a switch is in its own collision domain. 





* Access switches allow endpoint devices to connect, traffic is sent to and drom the endpoints is transmitted through these switches. 
* Distribution Layer switches allow traffic to passed between the access switches. 
* Core switches allow traffic to passed between distribution switches.
* Core switches are really there for speed
* it kind of is like A tree data structure. easier to go to the root and across.
