# 3.1 Graphs in the Machine Learning Applications


Machine Learning workflow (defined in Chapter 1 - with the CRISP-DM Model, can be decomposed to the following steps:

1. Select the data sources, gather data, and prepare the data.
2. Train the model (the learning phase).
3. Provide predictions.

** Some learning algos don't have a model. Instance Based algos don't have a seperate learning phase.We won't consider these for graphs. 






* The graph representation of data doesn't support only graph-based algorithims; it can feed multiple types of algos. Specifically, the graph representation is helpful for the following tasks:


1. Feature Selection: graphs are easy to query and therefore identifying input variables is easy. 
2. Data filtering - Easy to navigate relationships among objects make it easy to filter out useless data during the training phase
3. Data preperation - Graphs make it easy to clean data, removing spurious entries.
4. Data Formatting - easy to export data to which format is neccessary.

### 3.2.1 Monitoring a subject






* The graph modeling described here illustrates a graph construction technique. The resulting graph can be generalized as a co-occurrence graph. The nodes represent entities (in this case, cellular towers), and the relationships represent the fact that two entities belong to a common set or group. (In the CTN, the set is the row in the table indicating that at a specific point in time, the phone can reach two towers.) This technique is a powerful one thats used in many algorithms and machine learning applications as a data-preprocessing step before analysis.
