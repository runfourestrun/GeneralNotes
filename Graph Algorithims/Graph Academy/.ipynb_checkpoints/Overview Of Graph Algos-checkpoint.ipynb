### When not to use Graphs

* In general, very organized data that follows a regular schema, with few connections between data elements does not warrant a graph

#### Examples

* Questions with just a few or flat connections (not nested)
* Questions solved with specific, well crafted queries
* Simple statistical results (sum, averages, ratios)

### Monopartitie vs Bipartite Graphs





* Centrality and Clustering/Community Detection algos are primarily designed to run on monopartite graphs. Usually, it is a mistake to run centrality on a bipartie graph. To overcome this, you are able to infer a monopartite graph from a bipartite graph.

# Types of Algos




### Community Detection 

* Communitiy detecton algos are used to find clusters of communties that the nodes from in a network. They are also used to examine how tightly-knit some of these communities are. 

* Connected Components
* Label propogation
* Louvanian Modularity


### Centrality

* Used to find the most influential nodes and their role in a network based on graph topology.
* Used to infer group dynamics such as : credibility, rippling vulnerability, and bridges between groups

* PageRank



### Pathfinding and Search

* are usually used to find the shortest path between nodes in a netwrok. Used for evaluating routes such a s physical logistics and least-cost call or IP routing



### Similarity Algos

* used to find similar nodes in a network based on graph topology and their properties
* used for personalized recommendations and developing categorical hierarchies

* Jaccard Similarityy and Cossine Similarity algos. 


### Link Prediction

* Link prediction algos help determine the cliseness of a pair of nodes. They consider the proximity of the nodes, as well as structural elements, to predict unobserved or future relationships. 


### Node embeddings

* Node embeddings compute low-dimensional vector representations of nodes in a graph. Can be used in machine learning inputs
